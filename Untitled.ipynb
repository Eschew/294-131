```python
topics = glob.glob("/data/efros/ahliu/vot2013/*")
topics


files = glob.glob("/data/efros/ahliu/vot2013/singer/*")
jpg_files = [f for f in files if '.jpg' in f]

jpg_files.sort(key=lambda x: float(x.split("/")[-1].split(".jpg")[0]))

ground_truth = open("/data/efros/ahliu/vot2013/singer/groundtruth.txt")
boxes = []

for lines in ground_truth:
    boxes.append([float(i) for i in lines.split(",")])
```

```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt

fig2 = plt.figure()
ax2 = fig2.add_subplot(111, aspect='equal')

i = 1

ax2.imshow(skio.imread(jpg_files[0]))

ax2.add_patch(patches.Rectangle(
        (boxes[0][0], boxes[0][1]),
        boxes[0][2],
        boxes[0][3],
        fill=False      # remove background
    ))


plt.show()
```
